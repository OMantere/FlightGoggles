```python
def linear_interpolate(p, n):
    res = np.zeros((n * p.shape[0], p.shape[1]))
    ii = 0
    for i in range(p.shape[0] - 1):
        res[ii, :] = p[i, :]
        for j in range(n):
            ii += 1
            res[ii, :] = p[i, :] + (p[i + 1, :] - p[i, :]) * (1.0*j/n)
        ii += 1
    return res
```

```python
import yaml

def read_yaml(f):
    stream = open(f)
    docs = yaml.load_all(stream)
    dat = {}
    for doc in docs:
        for k, v in doc.items():
            dat[k] = v
    return dat
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.interpolate import splprep, splev

gate_dat = read_yaml('gate_locations_0.yaml')
pert_dat = read_yaml('../perturbations/perturbed_gates_0.yaml')
pert_dat = {k: np.array([np.float32(i) for i in v.strip().split(',')]) for k, v in pert_dat.items()}

init_pose = np.array([18.0, -23.0, 5.3])
gatess = ['Gate10', 'Gate21', 'Gate2', 'Gate13', 'Gate9', 'Gate14', 'Gate1', 'Gate22', 'Gate15', 'Gate23', 'Gate6']

p = np.zeros((len(gatess)+1, 3))
p[0, :] = init_pose
for i, g in enumerate(gatess):
    p[i+1, :] = np.mean(np.array(gate_dat[g]['location']), axis=0)
    p[i+1, :] -= pert_dat[g]
    
n = 10000
tck, u = splprep(p.T, u=None, s=0.0, per=0, k=1) 
u_new = np.linspace(u.min(), u.max(), n)
ev = np.array(splev(u_new, tck, der=0)).T
ev = linear_interpolate(p, n/10)
```

```python
# plot
axx = ['x', 'y', 'z']
def plot_gates(xx, yy):
    fig, ax = plt.subplots()
    plt.plot(ev[:,xx], ev[:,yy], 'b--')
    ax.scatter(p[:,xx], p[:,yy])
    for i in range(len(p)):
        ax.annotate(str(i+1), (p[i, xx], p[i, yy]))
    plt.xlabel(axx[xx])
    plt.ylabel(axx[yy])
```

```python
plot_gates(0, 1)
plot_gates(1,2)
plot_gates(0,2)
```
